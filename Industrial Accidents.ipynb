#             Industrial Safety and Health Analytics Database 

## The analysis of the dataset is to help companies understand why employees still suffer from injuries or accidents in plants. This might help the community explore and take better insights from the data. This can also help in manufacturing plants with the additional security and safety required.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
acc = pd.read_csv('IHMStefanini_industrial_safety_and_health_database.csv')
```

```python
acc.head()
```

```python
acc.info()
```

```python
acc.shape
```

### Let us check the datatype of the content of the dataframe. 

```python
ddict = pd.DataFrame(acc.dtypes)
ddict
```

### isnull() will help us check if there are any missing values in the dataframe. 

```python
ddict['Missing Values'] = acc.isnull().sum()
ddict
```

```python
ddict['Unique Values'] = acc.nunique()
ddict
```

```python
acc.describe()
```

### Insight :  As we can see with the help of this plot, country_01 has so far the highest count for accident                              levels.  

```python
sns.catplot(data=acc,kind='count',x='Countries',hue='Local',height=6,aspect=2)
```

### As the original dataset is predefined let us modify the column names according to our understanding.

```python
acc.rename(columns={'Data':'Date','Genre':'Gender','Employee ou Terceiro':'Employee Type','Risco Critico':'Critical Risks'},inplace=True)
acc
```

```python
acc['Date'] = pd.to_datetime(acc['Date'])
```

```python
acc.groupby('Date').count()['Local'].plot(figsize=(12,6))
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

### The above graph helps us understand the risks of the number of incidents that took place, date wise. The readings are pretty less on an average except certain times like around February to MARCH 2016 or that of 2017. 

```python
order = {'I':1,'II':2,'III':3,'IV':4,'V':5}
```

```python
fig = sns.FacetGrid(acc,height=4,palette='winter',hue='Gender',col='Industry Sector')
fig.map(sns.countplot,'Accident Level',order=order)
plt.legend(bbox_to_anchor=(1,1))
```

```python
acc_ind = acc.groupby('Industry Sector').count()['Date']
acc_ind.plot(kind='pie',figsize=(5.5,5.5),autopct='%.2f',title='Number of Incidents')
```

### The above pie chart tells us that maximum number of incidents take place in the mining sector. 

### Insight : More care should be taken in the mining sector as there might be certain factors due to which the                       workers might face problems.

```python
sns.countplot('Employee Type',data=acc)
```

### There is not much count difference between the third party and the employees but not much information has been provided regarding the workers.

### Insight : Matters like what kind of work the third party workers or the employees do should be looked into                         which might help reduce the graph.

```python
weekday={'0':'Monday','1':'Tuesday','2':'Wednesday','3':'Thursday','4':'Friday','5':'Saturday','6':'Sunday'}
w_order={'Monday':1,'Tuesday':2,'Wednesday':3,'Thursday':4,'Friday':5,'Saturday':6,'Sunday':7}
```

```python
week_d=acc.pivot_table(index='weekday', columns='Industry Sector', aggfunc='count')['Accident Level']
acc.groupby('weekday').count()['Accident Level'].loc[w_order].plot(kind='line',figsize=(9,4),xticks=range(0,8))
```

### Also, the accident levels on a weekday basis are usually higher on wednesdays and a noticeable drop exists on the weekends.

### Insight :  This also indicates the number of workers working in the plants during the week.

```python
n = np.nan
acc_table = acc.pivot_table(index='Critical Risks',columns='Industry Sector',aggfunc='count')['Date']
acc_table.replace(n,0,inplace=True)
acc_table['Total'] = acc_table.sum(axis=1)
acc_table
```

### The above table mentions the number of entries for the particular critical risks, the highest being the 'Others' category. Hence, let us temporarily drop the 'Others' row to get a clearer representation. 

### Insight : The 'Others' count being the highest it could also mean that the reporting or managing of the                               critical risks was not handled properly.

```python
acc_table.drop('Others',axis=0,inplace=True)
```

```python
acc_table.nlargest(10,'Total').plot(kind='bar', figsize=(15,5), cmap='summer')
```

### This graph specifies the top ten critical risks responsible for the existence of the accident levels.

### Insight : - The incident types in the metal sector are mostly pressed, manual tools, chemical substances,                            pressurized systems or chemical substances, etc.
###                - The frequent incident types in the mining sector are projections, vehicles and mobile                                              equipments, etc.
###                - Others include venomous animals, bees, fall, etc.

### So the overall project analysis specifies the need for safety and precaution measures to be taken in plants for the employees to improve quality of life, work, and employment. This could further help in the development of the industry sector also.
